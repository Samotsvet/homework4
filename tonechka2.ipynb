```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.interpolate as sc
from scipy.interpolate import lagrange as lag
import numpy.polynomial.polynomial as npp
```

```python
samoc = np.load('ladybug.npz')
```

```python
samoc.files
```

```python
x = samoc['arr_0']
y = samoc['arr_1']
```

```python
plt.plot(x, y)
plt.show()
```

```python
fun = lag(x, y)
```

```python
npp.Polynomial(fun).coef
```

```python
# Ну функция явно периодическая, причем в нуле ноль, а наклон в нуле 2, все четные кооф равны нулю. Вероятно
# это синус, проверим
```

```python
rew = np.arange(-10, 10, 0.001)
plt.plot(rew, np.sin(2*rew))
plt.plot(rew, fun(rew))
plt.plot(x, y)
plt.show()
```

```python
#Все совпадает очень хорошо, значит скорее всего это это sin(2x)
```
